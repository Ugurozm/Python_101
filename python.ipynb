BİHTER ZİYAGİL ÖRNEK


Tabi Siz Anneleri Tarafından Size Emanet Edilen Çocuklara Her Bakımdan Yetersiz Gördüğünüz Bir Kadının Annelik Etmesine Şiddetle Karşısınız Ama.'

Cümlesindeki "anne" kelimelerini sayacak bir döngü ve kelime sayısını söyleyecek bir fonksiyon yazın.

```python

cumle = "Tabi Siz Anneleri Tarafından Size Emanet Edilen Çocuklara Her Bakımdan Yetersiz Gördüğünüz Bir Kadının Annelik Etmesine Şiddetle Karşısınız Ama."
kelimeler = cumle.lower().split()
toplam = 0
for i in kelimeler:
    if "anne" in i:    # i de anne kelimesi geçiyorsa kabul eder   if i[:4] == "anne" kullanarak da ilk 4 harften kontrol yapabiliriz
        toplam += 1
print(f"Toplam {toplam}")
```

ÇANAKKALE KÖPRÜSÜ ÖRNEK2

1915 Çanakkale Köprüsü'nde saatte 112 + X*2 hızla gidiyorsun ve köprüden çıkışta ücret ile beraber hızına göre bir ceza ödeyeceksin. Otomobil ücretinin 272.5 TL olduğu durumda.

120 hız sınırının %0 - 10 üzerindeysen ücretin %50'si

%10 - 20 üstündeysen %75'i

%20'den fazlaysan %150'si kadar ceza ödeyeceksin.

Bu duruma göre hızına göre ödeyeceğin toplam miktarı sana söyleyen bir Fonksiyon yaz

Dipnot: X = şuanki saat (Örn: 19:15 -----> 112 + 19*2 = 150 km/h)

```python
def ucretHesaplama(saat):
    saat = saat - (saat % 1) #Saatin ondalıklı kısmını ortadan kaldırmaya ayarar
    hiz = 112 + saat*2

    if hiz >= 120 and hiz <= 120*1.1: #120 <= hiz <= 132:      
        ucret = 227.5 + 227.50 * 0.5
        print(f" Tutar: {ucret}")
    elif hiz > 120*1.1 and hiz < 120 * 1.2:
        ucret = 227.5 + 227.5 * 0.75
        print(f" Tutar: {ucret}")
    elif hiz >= 120 * 1.2:
        ucret = 227.5 + 227.5 * 1.5
        print(f" Tutar: {ucret}")
    else:
        print("Ödenecek ucret 227.5 ")
ucretHesaplama(3.85)        
```
